# Prisonser Dillemma

```python
from Prison import PrisonersDilemma
from Prisoner import Prisoner
import random
```

```python
rounds = 1
episodes = 10000
prison = PrisonersDilemma(rounds)
prisoner_a = Prisoner(rounds)
prisoner_b = Prisoner(rounds)

for episode in range(episodes):
    if episode % 1000 == 0:
        print(f"Episode {episode}")
    o = prison.reset()
    for i in range(rounds):
        action_a = prisoner_a.act(o[0])
        action_b = prisoner_b.act(o[1])
        o, r, f, _ = prison.step(action_a, action_b)
        prison.render()

        prisoner_a.train(o[0], r[0])
        prisoner_b.train(o[1], r[1])

        # print(f"Episode {episode + 1}, Round {i + 1}")
        print("Observation A: ", o[0])
        print("Observation B: ", o[1])
        print("Rewards: ", r)
        # print("Final: ", f)
        # print()

        if f:
            break
```

```python
# Test the trained agents
print("Testing the trained agents")
o = prison.reset()

score_a = 0
score_b = 0
for i in range(rounds):
    action_a = prisoner_a.act(o[0])
    action_b = prisoner_b.act(o[1])
    o, r, f, _ = prison.step(action_a, action_b)
    prison.render()

    score_a += r[0]
    score_b += r[1]

    print(f"Round {i + 1}")
    print("Observation A: ", o[0])
    print("Observation B: ", o[1])
    print("Rewards: ", r)
    print("Final: ", f)
    print()

    if f:
        break

print("Score A: ", score_a)
print("Score B: ", score_b)
```
