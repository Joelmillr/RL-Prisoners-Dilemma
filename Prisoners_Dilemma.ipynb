# Prisonser Dillemma

```python
from Prison import PrisonersDilemma
from Prisoner import q_table_prisoner
```

```python
rounds = 1
episodes = 10000
prison = PrisonersDilemma(rounds)
prisoner_a = q_table_prisoner()
prisoner_b = q_table_prisoner()

for episode in range(episodes):
    if episode % 1000 == 0:
        print(f"Episode {episode}")
    obs_a, obs_b = prison.reset()
    for i in range(rounds):
        act_a = prisoner_a.choose_action(obs_a)
        act_b = prisoner_b.choose_action(obs_b)
        _, obs_a, obs_b, rew_a, rew_b, f = prison.step(act_a, act_b)
        prison.render()

        prisoner_a.update(rew_a, obs_a)
        prisoner_b.update(rew_b, obs_b)

        # print(f"Episode {episode + 1}, Round {i + 1}")
        print(f"Observation A: {'B snitched' if obs_a else 'B stayed silent'}")
        print(f"Observation B: {'A snitched' if obs_b else 'A stayed silent'}")
        print("Reward A: ", rew_a)
        print("Reward B: ", rew_b)
        print()
```

```python
# Test the trained agents
print("Testing the trained agents")
o = prison.reset()

score_a = 0
score_b = 0
for i in range(rounds):
    action_a = prisoner_a.act(o[0])
    action_b = prisoner_b.act(o[1])
    o, r, f, _ = prison.step(action_a, action_b)
    prison.render()

    score_a += r[0]
    score_b += r[1]

    print(f"Round {i + 1}")
    print("Observation A: ", o[0])
    print("Observation B: ", o[1])
    print("Rewards: ", r)
    print("Final: ", f)
    print()

    if f:
        break

print("Score A: ", score_a)
print("Score B: ", score_b)
```
